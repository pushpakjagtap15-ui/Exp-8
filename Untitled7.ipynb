```python
# Study Of Basics Of NumPy
import numpy as np # Importing the NumPy
Array1 = np.array([[1,2,3],[4,5,6]])
print(Array1)

Dimension = Array1.ndim # Gives the dimensions of array
print(Dimension)

Shape = Array1.shape # Gives (Rows, Columns)
print(Shape)

Size = Array1.size # Gives the Total number of elements
print(Size)

Type = Array1.dtype # Gives the datatype of elements
print(Type)

Null_Matrix = np.zeros((2,3))
print(Null_Matrix)

Ones_Matrix = np.ones((2,3)) # Gives the unit matrix with specific order
print(Ones_Matrix)

Diagonal_Matrix = np.eye((7)) # Gives the diagonal matrix with specific order
print(Diagonal_Matrix)

Arrange = np.arange(1,10,2) # Arranging the elements
print(Arrange)

print(np.linspace(1,10,5))

print(Array1+2) # Each element of the array is Added by 2

Mean = np.mean(Array1) # Mean of the data in array
print(Mean)

Median = np.median(Array1) # Median of array data
print(Median)

MinValue = np.min(Array1) # Minimum value in data
print(MinValue)

MaxValue = np.max(Array1) # Maximum value in data
print(MaxValue)

Sum = np.sum(Array1) # Adding of the data in Array 1
print("Sum of the data in array is : ",Sum)
```
